```python
import json
import matplotlib.pyplot as plt

with open("mean_absolute_error.json") as f:
    data = json.load(f)

mae_percents = []
for session in data:
    # print(data[session])
    mae_percent =  data[session]['t2gen_vs_t2gt'] / data[session]['t1gt_vs_t2gt'] * 100
    mae_percents.append(mae_percent)

# fig1, ax1 = plt.subplots()
fig = plt.figure()
plt.boxplot(mae_percents)
fig.suptitle("Generated MAE vs. Identity MAE as a Percentage")
# plt.boxplot(mae_percents)
plt.xticks([1], [''])
plt.ylabel("Generated_MAE / Identify_MAE * 100")
fig.savefig("mae_boxplot.png", bbox_inches='tight')
```
